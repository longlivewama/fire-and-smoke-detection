```python
# Step 1: Import Libraries
import cv2
from ultralytics import YOLO

# Step 2: Define the detection function
def detect_on_video(input_path, output_path="output_detected.mp4", model_path="best.pt"):
    model = YOLO(model_path)
    cap = cv2.VideoCapture(input_path)

    if not cap.isOpened():
        print("âŒ Error opening video file.")
        return

    # Get video properties
    width  = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    fps    = cap.get(cv2.CAP_PROP_FPS)

    # Create output video writer
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')
    out = cv2.VideoWriter(output_path, fourcc, fps, (width, height))

    print("ğŸš€ Running YOLO detection...")
    while True:
        ret, frame = cap.read()
        if not ret:
            break

        # Inference and annotate
        results = model(frame)
        annotated_frame = results[0].plot()
        out.write(annotated_frame)

    cap.release()
    out.release()
    print(f"âœ… Done! Saved at: {output_path}")

# Step 3: Run detection
input_video_path = "/content/Untitled video - Made with Clipchamp (6).mp4"
output_video_path = "/content/annotated_output.mp4"
detect_on_video(input_video_path, output_video_path)
```
